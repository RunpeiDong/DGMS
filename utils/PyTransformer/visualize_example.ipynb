```python
import torch
import torch.nn as nn
import torchvision
import torchvision.models as models

from transformers.torchTransformer import TorchTransformer
```

```python
model_name = "alexnet"
model = models.__dict__[model_name]()
model.eval()
```

```python
input_tensor = torch.randn([1, 3, 224, 224])
```

## visualization

### without saving image

```python
transformer = TorchTransformer()
```

```python
dot = transformer.visualize(model, input_tensor = input_tensor)
```

```python
dot
```

### save image

```python
model.cuda()
input_tensor = input_tensor.cuda()
```

```python
dot = transformer.visualize(model, input_tensor = input_tensor, save_name = "examples/{}".format(model_name), graph_size = 80)
```
